```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.model_selection import cross_val_score 
```

```python
veri = pd.read_csv("breast-cancer-wisconsin.data")
veri.head()
```

```python
veri.replace('?', -7777, inplace=True)
veri.head(24)
```

```python
veri= veri.drop(['id'], axis=1)
veri.head()
```

```python
y=np.array(veri.benormal)
x=np.array(veri.drop(['benormal'], axis=1))
print("y:")
print(y)

print("\nx:")
print(x)
```

```python
imp = SimpleImputer(missing_values=-7777, strategy="mean")
x = imp.fit_transform(x)
print(x)
#soru işarerti olan yerlere 777 koymak yerine, 
#o satırın ortalamasını alıyo ki tahminler en az şekilde etkilensin
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
#hangi k değerinin daha iyi sonuç verdiğini görmek için ve tek olması gerek
for z in range(25):
    z = 2*z+1
    print("En yakın",z,"komşu kullandığımızda tutarlılık oranımız:")
    tahmin = KNeighborsClassifier(n_neighbors=z, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='euclidean', metric_params=None, n_jobs=1)
    tahmin.fit(x,y)
    ytahmin = tahmin.predict(x)

    basari = accuracy_score(y, ytahmin, normalize=True, sample_weight=None)
    print(basari)
```
